# Formatting LLM Output

```python
import torch
from transformers import pipeline

model_id = "unsloth/Llama-3.2-1B-Instruct"
pipe = pipeline(
    "text-generation",
    model=model_id,
    torch_dtype=torch.bfloat16,
    device_map="auto",
)
```

```python
# System message
system_message = "You are a helpful chatbot who can help solving all kinds of problems."
```

```python
# User message
user_message = """Translate the original sentence delimited in <<>> into German.
Give your final answer with a JSON format as follows:
{"original": <sentence to be translated>,
"translated": <translated sentence>}
<<Architecture is Frozen Music.>>
"""
```

```python
# Generate responce
messages = [
    {"role": "system", "content": system_message},
    {"role": "user", "content": user_message},
]
outputs = pipe(
    messages,
    max_new_tokens=256,
)
```

```python
result = outputs[0]["generated_text"][-1]["content"]

print(result)
```

```python
# Output from LLM as input for other functions
import json
results = json.loads(result)

results['original_length'] = len(user_message)
results['translated_length'] = len(results['translated'])

print(results)
```
